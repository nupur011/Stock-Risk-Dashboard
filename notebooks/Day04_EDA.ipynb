```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
stock=pd.read_csv("../data/reliance_cleaned.csv", index_col=0, parse_dates=True)
stock.head()
```

```python
stock.info()
```

```python
plt.figure(figsize=(12,6))
plt.plot(stock.index, stock['Close'])
plt.title('Reliance Stock Closing Price')
plt.xlabel('Year')
plt.ylabel('Closing Price')

plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(stock.index, stock['Volume'])
plt.title('Reliance Trading Volume')
plt.xlabel('Year')
plt.ylabel('Volume')

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.hist(stock['Close'], bins=30)
plt.title('Distribution of Close Prices')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.show()
```

Observations:

1. Reliance stock shows a long-term upward trend.

2. The highest closing price was 1581.824462890625 on 2024-07-08.

3. The lowest closing price was 168.7269287109375 on 2014-02-28.

4. Trading volume showed spikes during mid 2020.

5. Most closing prices are concentrated between 200 and 300.
