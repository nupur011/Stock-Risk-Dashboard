```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
stock=pd.read_csv("../data/reliance_features_v1.csv", index_col=0,parse_dates=True)
stock.head()
```

```python
stock['Tomorrow_Close']=stock['Close'].shift(-1)
```

```python
stock['Target']=np.where(stock['Tomorrow_Close'] > stock['Close'], 1, 0)
```

```python
stock[['Close', 'Tomorrow_Close', 'Target']].head(10)
```

```python
stock['Lag_1'] = stock['Return'].shift(1)
```

```python
stock['Lag_2'] = stock['Return'].shift(2)
```

```python
stock['Lag_3'] = stock['Return'].shift(3)
```

```python
stock[['Return', 'Lag_1', 'Lag_2', 'Lag_3']].head(10)
```

```python
stock.isnull().sum()
```

```python
stock.dropna(inplace=True)
```

```python
stock.isnull().sum()
```

```python
stock.to_csv("../data/reliance_m1_dataset.csv")
```
